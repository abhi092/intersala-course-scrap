```python
import pandas
import requests
from bs4 import BeautifulSoup
response =requests.get("https://internshala.com/internships")
content = response.content
soup = BeautifulSoup(content,"lxml")
top_rest = soup.find_all("div",attrs={"class": "table-cell"})
list_tr = soup.find_all("div",attrs={"class": "internship_meta"})
list_rest =[]
for tr in list_tr:
    dataframe ={}
    dataframe["NAME:"] = (tr.find("a")).text.replace('\n', '')
    dataframe["ADDRESS:"] = (tr.find("a",attrs={"class": "location_link"})).text.replace('\n', '')
    dataframe["PRICE"] = (tr.find("td",attrs={"class":"stipend_container_table_cell"})).text.replace('\n', '')
    

    
    #frames = [dataframe["NAME:"],dataframe["ADDRESS:"],dataframe["PRICE"]]
    list_rest.append(dataframe)

list_rest
df = pandas.DataFrame(list_rest)
df.to_csv("A",index=True)
print('--------------------------INTERNSHIPS---------------------------------------')
print(df)
```
